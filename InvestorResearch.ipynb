```python
import requests

def google_search(query, api_key, cse_id, num_results=10):
    url = f"https://www.googleapis.com/customsearch/v1"
    params = {
        'key': api_key,
        'cx': cse_id,
        'q': query,
        'num': num_results,
    }
    response = requests.get(url, params=params)
    return response.json()

def main():
    # Replace these with your actual API key and CSE ID

    # Investor's name
    investor_name = "Danielle Strachman"
    
    # Search queries based on the instructions
    queries = [
        f'{investor_name} investor',
        f'{investor_name} podcast',
        f'{investor_name} blog',
        f'{investor_name} interview',
        f'{investor_name} news',
    ]

    for query in queries:
        print(f"Searching for: {query}")
        results = google_search(query, api_key, cse_id)
        if 'items' in results:
            for item in results['items']:
                print(f"Title: {item['title']}")
                print(f"Link: {item['link']}")
                print(f"Snippet: {item['snippet']}")
                print()
        else:
            print("No results found.")
        print("="*40)


main()
```

```python
from submind.memory.memory import remember
from decouple import config
from langchain_openai import OpenAIEmbeddings
from langchain_pinecone  import PineconeVectorStore
from pinecone import Pinecone
import uuid 
from langchain_experimental.text_splitter import SemanticChunker
from langchain_text_splitters import RecursiveCharacterTextSplitter
import json
from urllib.parse import urlparse, urljoin
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
import re
from langchain.document_loaders import BSHTMLLoader
from langchain_chroma import Chroma
import requests
from langchain_anthropic import ChatAnthropic
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

knowledge = remember(submind._submind)

top_concern = """The top investor concern is whether Rivalz can effectively compete and gain market share in a highly competitive snack industry, given the established presence of major brands and emerging health-focused competitors."""

REWRITE_IN_VOICE_PROMPT = """You are a powerful submind that is used to rewrite reports in the voice of a specified brand.

Here's the report to rewrite: {report}

Here are the brand guidelines and voice to use: {brand}

Rewrite the report with the same information but in the style and voice of the specified brand. Only rewrite the report, don't explain what you changed.
"""

model_claude = ChatAnthropic(model_name="claude-3-5-sonnet-20240620",
                                 anthropic_api_key=config("ANTHROPIC_API_KEY"),
                                 anthropic_api_url="https://anthropic.hconeai.com/",
                                 max_tokens=4096,
                                 model_kwargs={
                                     "extra_headers": {
                                         "Helicone-Auth": f"Bearer {config('HELICONE_API_KEY')}",
                                         "Helicone-Property-Step": "Brand Voice rewrite",
                                         "Helicone-Property-UUID": "deckdocs.com"

                                     }
                                 }
                                 )
prompt = ChatPromptTemplate.from_template(REWRITE_IN_VOICE_PROMPT)
chain = prompt | model_claude | StrOutputParser()
new_report = chain.invoke({
    "report": top_concern,
    "brand": knowledge
})

print(new_report)
```
